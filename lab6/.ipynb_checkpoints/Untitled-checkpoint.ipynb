```python
import random
import numpy as np

n = 39
min_value = -199
max_value = 199
negative_percentage = 0.3

array = [random.randint(min_value, max_value) for _ in range(n)]

min_element = min(array)
array[random.randint(0, n-1)] = min_element
array[random.randint(0, n-1)] = min_element

neg_count = int(n * negative_percentage)
neg_indices = random.sample(range(n), neg_count)
for i in neg_indices:
    array[i] = -abs(array[i])

print("Згенерований масив:", array)
```

```python
with open("array.txt", "w") as file:
    for number in array:
        file.write(f"{number}\n")
```

```python
with open("array.txt", "r") as file:
    loaded_array = [int(line.strip()) for line in file]

print("Завантажений масив:", loaded_array)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
for i, value in enumerate(loaded_array):
    color = 'red' if value == min(loaded_array) else ('blue' if value < 0 else 'green')
    plt.scatter(i, value, color=color, s=50)
    if color == 'red':
        plt.text(i, value, f'{value}', fontsize=8, ha='center')

plt.title("Точкова діаграма масиву")
plt.xlabel("Індекс")
plt.ylabel("Значення")
plt.show()
```

```python
min_positions = [i for i, x in enumerate(loaded_array) if x == min(loaded_array)]
print("Кількість мінімальних елементів:", len(min_positions))
print("Позиції мінімальних елементів:", min_positions)
```

```python
neg_indices = [i for i, x in enumerate(loaded_array) if x < 0]
if len(neg_indices) > 1:
    segment = loaded_array[neg_indices[0] + 1 : neg_indices[1]]
    print("Кількість елементів між першим і другим від'ємними елементами:", len(segment))
    print("Сума елементів між першим і другим від'ємними елементами:", sum(segment))
else:
    print("Недостатньо від'ємних елементів для підрахунку")
```

```python
less_than_100 = [x for x in loaded_array if abs(x) <= 100]
greater_than_100 = sorted([x for x in loaded_array if abs(x) > 100], reverse=True)
sorted_array = less_than_100 + greater_than_100

print("Перетворений масив:", sorted_array)

plt.figure(figsize=(10, 5))
plt.plot(sorted_array, 'o-', color='purple')
plt.title("Перетворений масив")
plt.xlabel("Індекс")
plt.ylabel("Значення")
plt.show()
```
